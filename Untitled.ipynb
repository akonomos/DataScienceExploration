```python
from statsmodels.stats.power import tt_ind_solve_power
```

```python
#effect of different effect sizes
effect_size = 0.5
ratio = 1
nobel = tt_ind_solve_power(effect_size=effect_size,
                          nobs1=None,
                          alpha=0.05,
                          power=0.8,
                          ratio=ratio,
                          alternative='two-sided')
```

```python
print(f"The number of observations we need in group 1 is{nobel}")
```

```python
nobs2 = nobel*ratio
print(f"Sample size of group 2 is {nobs2}")
```

```python
print(f"The total sample size is {nobel+nobs2}")
```

```python
#t test analysis
import numpy as np
from scipy.stats import ttest_ind

np.random.seed(42)
treatment = np.random.normal(loc=0.1, scale=0.02, size=1000)
control = np.random.normal(loc=0.12, scale=0.02, size=1000)
treatment
```

```python
t, p = ttest_ind(treatment, control)
#provides t statistic and p value
#if p < 0.05, reject null hypothesis
if p<0.05:
    print('reject the null, there is a significant difference between treatment and control')
else:
    print('there is no difference between treatment and control')
```

```python
#diff in diff
import pandas as pd
from statsmodels.formula.api import ols
```

```python
np.random.seed(42)
df=pd.DataFrame({
    'time_period': np.repeat(['Before', 'After'], 100),
    'group': np.tile(['Control', 'Treatment'], 100),
    'outcome': np.concatenate([
        np.random.normal(10,2,100),
        np.random.normal(12,2,100),
    ])
})

df

df.loc[(df['group']=='Treatment') & (df['time_period']=='After'), 'outcome'] = df['outcome'] *1.2
df

```

```python
#diff n diff
model = ols('outcome ~ group * time_period', data=df).fit()
model.summary()
```
